# Uniform Distribution

#### All the outcomes have an equal probability of occurrence and mutually exclusive.
#### Probability of occurrence is uniformly distributed.

That means there is ***no bias***—no outcome is more likely than another within the specified set.

### Types of uniform distribution

#### 1. Discrete Uniform Distribution:

Can take a finite number of values and each value has equal probability of selection. Ex.  rolling a die (each number from 1 to 6 has an equal chance: 1/6).

#### 2. Continuous Uniform Distribution:  

Can take any value within a specified range. All values within this range are equally likely, but since there are infinitely many values, we use a probability density function (PDF) instead of direct probabilities. Ex. Choosing a random number between 0 and 1 and temperature o a particular day between some numbers.

### Continuous Uniform Distribution

#### Problem statement¶
 IT industry records the amount of time a software engineer needs to fix a  bug in the initial phase of software development in 'debugging.csv'.

Let

X = Time needed to fix bugs

X is a continuous random variable. Let's see the distribution of X and answer the below questions.

1. Find the probability that a randomly selected software debugging requires less than three hours.

2. Find the probability that a randomly selected software debugging requires more than two hours.

3. Find the 50th percentile of the software debugging time.

```python
# import the important packages
import pandas as pd  # library used for data manipulation and analysis
import numpy as np  # library used for working with arrays
import matplotlib.pyplot as plt  # library for plots and visualizations
import seaborn as sns  # library for visualizations
%matplotlib inline

import scipy.stats as stats  # this library contains a large number of probability distributions as well as a growing library of statistical functions
```

```python
debug = pd.read_csv('debugging.csv')
debug.head()
```

Let's plot the histogram of data along with the PDF of uniform distribution using the parameters minimum time required and maximum time required for bug fixing.

#### Histogram to view distribution

```python
plt.hist(debug['Time Taken to fix the bug'], density=True)
plt.axhline(1/4, color='r')
plt.title('Data Distribution')
plt.xlabel('Time required to fix the bug')
plt.ylabel('Probability')
plt.show()
```

#### Distribution plot of seaborn to see distribution

```python
sns.displot(debug['Time Taken to fix the bug'], kde=True)
plt.show()
```

Values between 1 and 5 have equal probability and endpoints are 1 and 5.
X~U(1,5)

```python
# import required libraries
from scipy.stats import uniform
```

```python
# Generate uniform distribution using uniform.pdf()
x = np.linspace(1,5,50)
probs = uniform.pdf(x, loc=1, scale=4)
probs
```

#### Q1. Find the probability that a randomly selected software debugging requires less than three hours.

 Less than 3 hours means CDF to get the area under 3, P(X<=3)


#### Plot the distribution

```python
x1 = np.linspace(1,3,25)
plt.plot(x,probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1, loc=1, scale=4), color='r')
plt.xlabel('Time required for bug fixing')
plt.ylabel('Probability')
plt.title('Continuous Uniform Distribution: X ~ U (1,5)')
plt.show()
```

```python
uniform.cdf(x=3, loc=1, scale=4)
```

### Q2. Find the probability that a randomly selected software debugging requires more than two hours.

#### P(X>2) 

#### Plot the distribution

```python
x1 = np.linspace(2,5,20)
plt.plot(x, probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1, loc=1, scale=4), color='r')
plt.xlabel('Time required for bug fixing')
plt.ylabel('Probability')
plt.title('Continuous Uniform Distribution: X ~ U(1,5)')
plt.show()
```

```python
# Area below 2 is calculated using CDF, P(X<2), for area above 2 i.e P(X>2) = 1- P(X<2)
1 - uniform.cdf(x=2, loc=1, scale=4)
```

#### Q3. Find the 50th percentile of the software debugging time.

For 50th percentile we use ppf(), PErcentie point function which is inverse of cdf(), ppf() gives the point in return of percentile.

```python
uniform.ppf(q=0.50, loc=1, scale=4)
```

##### Conclusion:
There is a 50% chance that a randomly selected software debugging requires less than three hours.

There is a 75% chance that a randomly selected software debugging requires more than two hours.

The 50th percentile of the software debugging time is 3 hours.
