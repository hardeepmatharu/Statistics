# Central Limit Theorem

## Sampling

A sampling technique where every item in the population has an equal chance of being selected.

***Sampling Distribution:*** 
The sampling distribution of a statistics is the probability distribution of that statistic when we draw many samples.

### Central Limit Theorem
The sampling distribution of the sample 'means' will approach normal distribution as the sample size gets bigger, no matter what the shape of the population distribution is.

#### Assumptions
1. Data must be randomly sampled.
2. Sample values must be independent of each other.
3. Samples should come from the same distribution.
4. Sample size must be sufficiently large(>=30), large sample size provides better estimate of the population mean. 

#### Population distribution is Uniform

```python
# import the important packages
import pandas as pd  # library used for data manipulation and analysis
import numpy as np  # library used for working with arrays
import matplotlib.pyplot as plt  # library for plots and visualizations
import seaborn as sns  # library for visualizations
%matplotlib inline

from scipy.stats import uniform  # this library contains a large number of probability distributions as well as a growing library of statistical functions
```

```python
# setting the seed for reproducibility
np.random.seed(1)
# creating a uniform distribution population of size 100000
uniform_pop = uniform.rvs(0,10,size=100000)
plt.hist(uniform_pop)
plt.title("Uniform Distribution Population")
plt.xlabel("X~U(0,10)")
plt.ylabel("Count")
plt.show()
```

Let's create sampling distribution from this population(sample size=5, number of samples=500)

```python
np.random.seed(1)
n=5
sample_means = []

for j in range(500):
    # draw the sample of size n
    sample = np.random.choice(uniform_pop, size=n)
    # calcualte the sample mean
    sample_mean = np.mean(sample)
    # append the sample mean to the sample means list
    sample_means.append(sample_mean)
# plot the histogram
sns.displot(sample_means, kde=True)
plt.title('Distribution of Sample Means for n = ' + str(n))
plt.xlabel('Sample Mean')
plt.ylabel('Count')
plt.show()
```

## Normal Distribution Population

```python
from scipy.stats import norm

np.random.seed(1)

normal_pop = norm.rvs(0,1,size=100000)
plt.hist(normal_pop, 200)
plt.title('Normal Distribution Population')
plt.xlabel('X~N(0,1)')
plt.ylabel('Count')
plt.show()
```

#### Let's create a sampling distribution from this population (sample size=5, number of samples = 500)

```python
# set the seed for reproducibility
np.random.seed(1)
# set the sample size to 5
n = 5
# list to store sample means
sample_means = []
# iterate the loop to draw multiple sample
for j in range(500):
  # draw a sample of size n
  sample = np.random.choice(normal_pop, size = n)
  # calculate the sample mean
  sample_mean = np.mean(sample)
  # append the sample mean to the sample_means list
  sample_means.append(sample_mean)
# plot the histogram of sample means.
sns.displot(sample_means, kde = True)
plt.title('Distribution of Sample Means for n = ' + str(n))
plt.xlabel('sample mean')
plt.ylabel('count')
plt.show()
```
